```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
# Loading the dataset
file_path = "/content/sample_data/annual-death-rates-in-different-age-groups-by-sex.csv"
data = pd.read_csv(file_path)

# Displaying the first five rows
data.head()
```

1. **Bar Plot** to visualize the Death Rate Comparison by Sex for ages 0 and 25, in the years 1800, 1900, and 2000 respectively.

```python
selected_years = data[data['Year'].isin([1800, 1900, 2000])]
fig, ax = plt.subplots(figsize=(10, 6))
width = 3.0
years = selected_years['Year']
ax.bar(years - width, selected_years['Central death rate - Type: period - Sex: female - Age: 0'], width=width, label='Female - Age 0')
ax.bar(years, selected_years['Central death rate - Type: period - Sex: male - Age: 0'], width=width, label='Male - Age 0')
ax.bar(years + width, selected_years['Central death rate - Type: period - Sex: female - Age: 25'], width=width, label='Female - Age 25')
ax.bar(years + 2 * width, selected_years['Central death rate - Type: period - Sex: male - Age: 25'], width=width, label='Male - Age 25')

ax.set_xlabel('Year')
ax.set_ylabel('Death Rate')
ax.set_title('Death Rate Comparison by Age and Sex (Years 1800, 1900, and 2000)')
ax.set_xticks([1800, 1900, 2000])
ax.legend()
plt.show()
```

2. **Line Plot** to visualize the Death Rate Trend for Females, for ages 0 and 25, in the years 1800, 1900, and 2000 respectively.

```python
filtered_data = data[data['Year'].isin([1800, 1900, 2000])]
plt.figure(figsize=(8, 5))
plt.plot(filtered_data['Year'], filtered_data['Central death rate - Type: period - Sex: female - Age: 0'], marker='o', label='Female - Age 0', color='purple')
plt.plot(filtered_data['Year'], filtered_data['Central death rate - Type: period - Sex: female - Age: 25'], marker='o', label='Female - Age 25', color='orange')

plt.xlabel('Year')
plt.ylabel('Death Rate')
plt.title('Death Rate Trend for Females (Ages 0 and 25) in 1800, 1900, and 2000')
plt.xticks([1800, 1900, 2000])
plt.legend()
plt.show()
```

3. **Scatter Plot** to visualize the Death Rate comparison by sex, for ages 10, 25 and 80, in the year 2000.

```python
filtered_data_2000 = data[data['Year'] == 2000]
plt.figure(figsize=(10, 6))
plt.scatter(['Age 10 Female', 'Age 10 Male',
             'Age 25 Female', 'Age 25 Male',
             'Age 80 Female', 'Age 80 Male'],
            [filtered_data_2000['Central death rate - Type: period - Sex: female - Age: 10'].values[0],
             filtered_data_2000['Central death rate - Type: period - Sex: male - Age: 10'].values[0],
             filtered_data_2000['Central death rate - Type: period - Sex: female - Age: 25'].values[0],
             filtered_data_2000['Central death rate - Type: period - Sex: male - Age: 25'].values[0],
             filtered_data_2000['Central death rate - Type: period - Sex: female - Age: 80'].values[0],
             filtered_data_2000['Central death rate - Type: period - Sex: male - Age: 80'].values[0]],
            color=['blue', 'blue', 'green', 'green', 'orange', 'orange'], s=100)

plt.xlabel('Group')
plt.ylabel('Death Rate')
plt.title('Death Rates for Ages 10, 25, and 80 (Male vs Female) in 2000')
plt.grid(True)
plt.show()
```

4. **Histogram** to visualize the Death Rate comparison by sex, for age 0, in the year 2000.

```python
filtered_data = data[(data['Year'] >= 1850) & (data['Year'] <= 1900)]
plt.figure(figsize=(12, 6))
plt.hist(filtered_data['Central death rate - Type: period - Sex: female - Age: 0'],
         bins=10, alpha=0.5, label='Female - Age 0', color='blue', edgecolor='black')
plt.hist(filtered_data['Central death rate - Type: period - Sex: male - Age: 0'],
         bins=10, alpha=0.5, label='Male - Age 0', color='red', edgecolor='black')

plt.xlabel('Death Rate')
plt.ylabel('Frequency')
plt.title('Histogram of Death Rates for Age 0 (1850-1900) by Sex')
plt.legend()
plt.show()
```

5. **Bar Plot** to visualize the Death Rate comparison by sex, for ages 0, 25 and 65, in the year 2020.

```python
data_2000 = data[data['Year'] == 2020]
ages = ['Central death rate - Type: period - Sex: female - Age: 0', 'Central death rate - Type: period - Sex: male - Age: 0',
        'Central death rate - Type: period - Sex: female - Age: 25', 'Central death rate - Type: period - Sex: male - Age: 25',
        'Central death rate - Type: period - Sex: female - Age: 65', 'Central death rate - Type: period - Sex: male - Age: 65']
age_labels = ['Female Age 0', 'Male Age 0', 'Female Age 25', 'Male Age 25', 'Female Age 65', 'Male Age 65']
death_rates = [data_2000[col].values[0] for col in ages]

plt.figure(figsize=(12, 6))
plt.bar(age_labels, death_rates, color=['purple', 'blue', 'purple', 'blue', 'purple', 'blue'])
plt.xlabel('Sex and Age Group')
plt.ylabel('Death Rate')
plt.title('Death Rate Comparison at Ages 0, 25, and 65 by Sex in 2020')
plt.show()
```

6. **Pie Plot** to visualize the Female Death Rates for ages 0, 15, 45 and 80.

```python
male_data = data[['Central death rate - Type: period - Sex: female - Age: 0',
                  'Central death rate - Type: period - Sex: female - Age: 15',
                  'Central death rate - Type: period - Sex: female - Age: 80',
                  'Central death rate - Type: period - Sex: female - Age: 45']]

total_death_rates = male_data.sum()
plt.figure(figsize=(10, 8))
plt.pie(total_death_rates, labels=total_death_rates.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Female Death Rates by Age Group')
plt.axis('equal')
plt.show()
```

7. **Pie Plot** to visualize the Male Death Rates for ages 0, 15, 45 and 80.

```python
male_data = data[['Central death rate - Type: period - Sex: male - Age: 0',
                  'Central death rate - Type: period - Sex: male - Age: 15',
                  'Central death rate - Type: period - Sex: male - Age: 80',
                  'Central death rate - Type: period - Sex: male - Age: 45']]
total_death_rates = male_data.sum()
plt.figure(figsize=(10, 8))
plt.pie(total_death_rates, labels=total_death_rates.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Male Death Rates by Age Group')
plt.axis('equal')
plt.show()
```

8. **Bar Plot** to visualize the Death Rates by Gender for Age 65 in randomly selected countries.

```python
import random
random_countries = random.sample(data['Entity'].unique().tolist(), 4)
filtered_data = data[
    (data['Entity'].isin(random_countries)) &
    (data['Year'] == data['Year'].max())
][['Entity', 'Central death rate - Type: period - Sex: female - Age: 65',
   'Central death rate - Type: period - Sex: male - Age: 65']]
filtered_data.columns = ['Country', 'Female Death Rate (Age 65)', 'Male Death Rate (Age 65)']
fig, ax = plt.subplots(1, 2, figsize=(12, 6), sharey=True)
ax[0].barh(filtered_data['Country'], filtered_data['Female Death Rate (Age 65)'], color='purple')
ax[0].set_title('Female Death Rate (Age 65)')
ax[0].set_xlabel('Death Rate')
ax[1].barh(filtered_data['Country'], filtered_data['Male Death Rate (Age 65)'], color='blue')
ax[1].set_title('Male Death Rate (Age 65)')
ax[1].set_xlabel('Death Rate')

plt.suptitle('Death Rates by Gender for Age 65 in Selected Countries')
plt.tight_layout()
plt.show()
```

9. **Line Plot** to visualize the Death Rates by Gender for Age 45 in Japan.

```python
country = 'Japan'
age_group = 45
filtered_data = data[data['Entity'] == country][[
    'Year',
    f'Central death rate - Type: period - Sex: female - Age: {age_group}',
    f'Central death rate - Type: period - Sex: male - Age: {age_group}'
]]
filtered_data.columns = ['Year', 'Female Death Rate (Age 45)', 'Male Death Rate (Age 45)']
plt.figure(figsize=(12, 6))
plt.plot(filtered_data['Year'], filtered_data['Female Death Rate (Age 45)'], label='Female', color='orchid', marker='o')
plt.plot(filtered_data['Year'], filtered_data['Male Death Rate (Age 45)'], label='Male', color='skyblue', marker='o')

plt.xlabel('Year')
plt.ylabel('Death Rate')
plt.title(f'Death Rate Trend for Age {age_group} in {country} (Female vs Male)')
plt.legend()

plt.grid(True)
plt.tight_layout()
plt.show()
```

10. **Scatter Plot** to visualize the Death Rates by Gender, for Age 0, in India.

```python
country = 'India'
age_group = 0

filtered_data = data[data['Entity'] == country][[
    'Year',
    f'Central death rate - Type: period - Sex: female - Age: {age_group}',
    f'Central death rate - Type: period - Sex: male - Age: {age_group}'
]]
filtered_data.columns = ['Year', 'Female Death Rate (Age 0)', 'Male Death Rate (Age 0)']
plt.figure(figsize=(10, 6))
plt.scatter(filtered_data['Year'], filtered_data['Female Death Rate (Age 0)'],
            color='orchid', s=60, label='Female (Age 0)')
plt.scatter(filtered_data['Year'], filtered_data['Male Death Rate (Age 0)'],
            color='skyblue', s=60, label='Male (Age 0)')
plt.xlabel('Year')
plt.ylabel('Death Rate')
plt.title(f'Comparison of Male and Female Death Rates at Age 0 in {country} (All Available Years)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
